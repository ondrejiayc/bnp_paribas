```python
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
data.info()
```

```python
data.describe()
```

```python
compData = data.dropna() # complicated way: [~pd.isnull(data).any(1)], data with no NaNs
```

```python
keys = set(compData.keys())

keys.remove('ID')
keys.remove('target')

compData2 = compData[list(keys)]
```

```python
compData2.select_dtypes(include=['object']).describe()
```

```python
# pandas can do categorical into integer - pd.get_dummies(variable), but it creates a new column for each category

from sklearn import preprocessing as prep

_frame = pd.DataFrame(compData2.select_dtypes(include=['object'])).apply(prep.LabelEncoder().fit_transform)

for col in compData2:
    try:
        compData2.loc[:,col] = _frame[col]
    except KeyError:
        pass
```

```python
compData2.head()
```

```python
cData2_sc = prep.scale(compData2)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)
```

```python
cDataTransformed = pca.fit_transform(cData2_sc)
```

```python
cDataTransformed.shape
```

```python
np.cumsum(pca.explained_variance_ratio_)
```
